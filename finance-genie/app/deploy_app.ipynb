# Deploy Graph-Enriched Lakehouse Explorer

This notebook creates and deploys the **Graph-Enriched Lakehouse Explorer** Streamlit app to Databricks Apps using the Databricks CLI.

The app source code lives at:
```
/Workspace/Users/ryan.knight@neo4j.com/graph-enriched-lakehouse/finance-genie/app/
```

**What this notebook does:**
1. Creates a new Databricks App
2. Deploys the Streamlit source code
3. Verifies the deployment

**After running this notebook**, you must add a SQL Warehouse resource via the UI (Step 4).

## Step 1: Configuration

Set the app name and source path. Change `APP_NAME` if you want a different app identifier.

````python
APP_NAME = "graph-lakehouse-explorer"
SOURCE_PATH = "/Workspace/Users/ryan.knight@neo4j.com/graph-enriched-lakehouse/finance-genie/app"

print(f"App name:    {APP_NAME}")
print(f"Source path: {SOURCE_PATH}")
````

## Step 2: Create the App

Create the Databricks App. **Run this once only** — it will error if the app already exists, which is safe to ignore.

````python
%sh
databricks apps create graph-lakehouse-explorer
````

## Step 3: Deploy

Upload the source code and start the app. This reads `app.yaml`, `app.py`, `backend.py`, `pages/`, and `requirements.txt` from the source path.

````python
%sh
databricks apps deploy graph-lakehouse-explorer \
  --source-code-path /Workspace/Users/ryan.knight@neo4j.com/graph-enriched-lakehouse/finance-genie/app
````

## Step 4: Add SQL Warehouse Resource

⚠️ **This step must be done in the Databricks UI — it cannot be automated via CLI.**

1. Navigate to **Apps** in the left sidebar
2. Click on **graph-lakehouse-explorer**
3. Go to **Settings** (or click the gear icon)
4. Click **+ Add resource**
5. Select **SQL Warehouse**
6. Choose your warehouse and set permission to **CAN_USE**
7. Set the resource key to `sql-warehouse` (must match the `valueFrom` in `app.yaml`)

The app will restart automatically after adding the resource.

## Step 5: Verify Deployment

Check the app status and retrieve the URL. The `url` field in the output is your live app link.

````python
%sh
databricks apps get graph-lakehouse-explorer
````

````python
%sh
# View logs if the app fails to start or shows errors
databricks apps logs graph-lakehouse-explorer
````

## Redeployment

After making code changes to the app source files, simply **re-run Step 3** (the deploy cell above). The CLI will upload the updated source and restart the app.

```bash
# That's it — just re-run the deploy cell, or from a terminal:
databricks apps deploy graph-lakehouse-explorer \
  --source-code-path /Workspace/Users/ryan.knight@neo4j.com/graph-enriched-lakehouse/finance-genie/app
```

No need to delete or recreate the app. Resources (SQL Warehouse) persist across redeployments.
